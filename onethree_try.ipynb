```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

```

```python
# Convert 'AGE' column
age_mapping = {
    '21 to 35': 1,
    '36 to 50': 2,
    '51 to 65': 3,
    '65+': 4
}
been['AGE'] = been['AGE'].map(age_mapping)
```

```python
# Convert 'GENDER' column
gender_mapping = {
    'Female': 1,
    'Male': 0
}
been['GENDER'] = been['GENDER'].map(gender_mapping)
```

```python
# Display unique values in a specific column
unique_values = been['JOB_ROLE'].unique()
print(unique_values)
```

```python
# Convert 'job role' column
job_mapping = {
'Business Analyst':0, 'Bank Teller':1, 'Loan Processor':2, 'Investment Banker':3,
 'Finance Analyst':4, 'Risk Analyst':5, 'Credit Analyst':6, 'Customer Service':7,
 'Human Resources':8, 'Mortgage Consultant':9}
been['JOB_ROLE'] = been['JOB_ROLE'].map(job_mapping)
```

```python
x = been.drop(['WORK_LIFE_BALANCE_SCORE'], axis=1)
x.head()
```

```python
y = been['WORK_LIFE_BALANCE_SCORE']
y.head()
```

```python
 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
# Train the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
predicted_value = model.predict([[5,2,6,9,1,5,3,3,2,3,2,6,2,6,4,2,2,4,1]])
actual_value = been.loc[1,'WORK_LIFE_BALANCE_SCORE']
print(predicted_value, actual_value)
# predictions = model.predict([[21,1], [22,0]])
# predictions
```

```python
# Evaluate the model
y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)
print(f'R Score: {r2}')
```
